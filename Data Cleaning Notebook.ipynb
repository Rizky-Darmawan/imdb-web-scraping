```python
data.loc[data['Aspect Ratio'] == '2.40:1', 'Aspect Ratio'] = '2.39 : 1'
data.loc[data['Aspect Ratio'] == '2:1', 'Aspect Ratio'] = '2 : 1'
data.loc[data['Aspect Ratio'] == '1.78 : 1 / (high definition)', 'Aspect Ratio'] = '1.78 : 1'
data.loc[data['Aspect Ratio'] == '17:9', 'Aspect Ratio'] = '17 : 9'
data.loc[data['Aspect Ratio'] == '4:3', 'Aspect Ratio'] = '4 : 3'
```

```python
data.head()
```

```python
data['User Reviews'] = data['User Reviews'].str.replace('.', '')
data['User Reviews'] = data['User Reviews'].str.replace('K', '00')
```

```python
data.head()
```

```python
data['User Reviews'] = pd.to_numeric(data['User Reviews'], errors = 'coerce')
```

```python
data['User Reviews'] = data['User Reviews'].fillna(0).astype('int64')
```

```python
data['Budget'] = data['Budget'].str.replace(',', '')
data['Budget'] = data['Budget'].str.replace('$', '')

data['US Gross'] = data['US Gross'].str.replace(',', '')
data['US Gross'] = data['US Gross'].str.replace('$', '')

data['Worldwide Gross'] = data['Worldwide Gross'].str.replace(',', '')
data['Worldwide Gross'] = data['Worldwide Gross'].str.replace('$', '')
```

```python
data.head()
```

```python
data['Budget'] = pd.to_numeric(data['Budget'], errors = 'coerce')
data['Budget'] = data['Budget'].fillna(0).astype('int64')

data['US Gross'] = pd.to_numeric(data['US Gross'], errors = 'coerce')
data['US Gross'] = data['US Gross'].fillna(0).astype('int64')

data['Worldwide Gross'] = pd.to_numeric(data['Worldwide Gross'], errors = 'coerce')
data['Worldwide Gross'] = data['Worldwide Gross'].fillna(0).astype('int64')
```

```python
data.dtypes
```

```python
data['Runtime_h'] = data['Runtime'].str.split(' ').str[0]
data['Runtime_m'] = data['Runtime'].str.split(' ').str[1]
```

```python
data.head()
```

```python
data['Runtime_h'] = data['Runtime_h'].str.replace('h', '')
```

```python
data['Runtime_h'] = data['Runtime_h'].fillna(0).astype('int64')
```

```python
data['Runtime_h'] = data['Runtime_h'] * 60
```

```python
data['Runtime_m'] = data['Runtime_m'].str.replace('min', '')
```

```python
data['Runtime_m'] = data['Runtime_m'].fillna(0).astype('int64')
```

```python
data.head()
```

```python
data.columns
```

```python
runtime_list = ['Runtime_h', 'Runtime_m']
data['Runtime_mins'] = data[runtime_list].sum(axis = 1)
```

```python
data.head()
```

```python
data.drop(['Runtime_h', 'Runtime_m'], axis = 1, inplace = True)
```

```python
data.head()
```

```python
data['Metascores'] = data['Metascores'].fillna(0)
data['Critic Reviews'] = data['Critic Reviews'].fillna(0)
```

```python
data['Metascores'] = data['Metascores'].astype('int64')
data['Critic Reviews'] = data['Critic Reviews'].astype('int64')
```
